```python
import tensorflow as tf
import numpy as np
```

```python
# 定义变量 X、Y，使用 numpy 随机生成 100 个随机数（ndarray）
x_data = np.random.rand(100)
y_data = x_data * 0.1 + 0.2
# 打印
print(type(x_data))
print(x_data[:10])
print(y_data[:10])
```

```python
# 定义一个线性模型
k = tf.Variable(0.)
b = tf.Variable(0.)
y = k * x_data + b
```

```python
# 优化线性模型
# 
# 定义损失函数
# 
# tf.reduce_mean() 求平均，tf.square() 求平方
loss = tf.reduce_mean(tf.square(y_data - y))

# 定义一个梯度下降法优化器，来训练模型
optimizer = tf.train.GradientDescentOptimizer(0.2)

# 最小代价函数
train = optimizer.minimize(loss)
```

```python
# 初始化图中变量
init_op = tf.global_variables_initializer()
```

```python
# 创建会话
with tf.Session() as sess:
#     在会话中，运行声明变量
    sess.run(init_op)
    for step in range(400):
        sess.run(train)
        if step % 10 == 0:
#             print(step, sess.run([k, b]))
            print('迭代次数%d'%(step), sess.run([k, b, loss]))
            
```
